```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Title')
plt.plot(x,y)
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('Title')
plt.plot(x,y)
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Title')
plt.plot(x,y)
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax1.set_xlabel('X')
ax1.set_ylabel('Z')
ax2.set_xlim([0,100])
ax2.set_ylim([0,1000])
plt.plot(x,z)
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')
ax2.set_xlim([20.0,22.0])
ax2.set_ylim([30,50])
plt.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,lw=3,ls='--')
axes[1].plot(x,z,color='r',lw=3)
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2,figsize=(12,2))
axes[0].plot(x,y,lw=3,ls='--')
axes[1].plot(x,z,color='r',lw=3)
```
